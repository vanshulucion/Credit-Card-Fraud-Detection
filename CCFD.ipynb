training and testing data bifurcation

```python
# splitting the data into training and testing sets
xTest,xTrain, yTest, yTrain = train_test_split(xData, yData, test_size = 0.2, random_state= 42)
print(len(xTest))
print(len(xTrain))


```

building random forest classifier using scikit learn

```python
from sklearn.ensemble import RandomForestClassifier
# random forest model creation
rfc = RandomForestClassifier()
rfc.fit(xTrain, yTrain)
# predictions
yPred = rfc.predict(xTest)
```

Building evaluating parameters- accuracy, recall, precision, F1 score for checking model accuaracy

```python
# importing and printing every score of the classifier
from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix,matthews_corrcoef
outliers = len(fraud)
outliers
errors = (yPred != yTest).sum()
errors

acc = accuracy_score(yPred, yTest)
print("Accuracy score is {}". format(acc))
recall = recall_score(yPred, yTest)
print("Recall score is {}". format(recall))
prec = precision_score(yPred, yTest)
print("Precision score is {}".format(prec))
f1 = f1_score(yPred, yTest)
print("F1 score is {}". format(f1))
matCorr = matthews_corrcoef(yPred, yTest)
print("Mattews correlation coefficient is {}". format(matCorr))
```

Building Confusion Matrix

```python
conMat = confusion_matrix(yPred, yTest)
conMat
plt.figure()
sns.heatmap(conMat, xticklabels= ['Valid','Fraud'], yticklabels= ['Valid', 'Fraud'], annot = True, cmap = 'Blues', fmt = 'd')
plt.title("Confusion Matrix for Random Forest Classifier")
plt.xlabel("True class")
plt.ylabel("Predicted class")
plt.show()



# # printing the confusion matrix
# LABELS = ['Normal', 'Fraud']
# conf_matrix = confusion_matrix(yTest, yPred)
# plt.figure(figsize =(12, 12))
# sns.heatmap(conf_matrix, xticklabels = LABELS,
#             yticklabels = LABELS, annot = True, fmt ="d");
# plt.title("Confusion matrix")
# plt.ylabel('True class')
# plt.xlabel('Predicted class')
# plt.show()
```
